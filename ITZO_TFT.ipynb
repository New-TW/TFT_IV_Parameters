```
read = '/content/drive/My Drive/Labwork/as depo.csv'
x = pd.read_csv(read)

gateV = pd.DataFrame(x, columns=['Gate Vg (V)'])
drainI = pd.DataFrame(x, columns=['Drain Id (A)'])
k=np.array(drainI)
print(max(k)/min(k)) ## on/off current ration
plt.title("as Depo",fontsize=25)
plt.xlabel("Gate V")
plt.ylabel("Drain I")
plt.plot(gateV,drainI)
y=np.array(drainI)
z=[]
for i in range(len(y)-1):
  z.append(2*(y[i]-y[i-1]))
t=np.array(gateV)
u=max(z)*(t-t[z.index(max(z))])+y[z.index(max(z))]
plt.ylim(0,max(y))
plt.plot(t,u)
plt.show()
th=-y[z.index(max(z))]/max(z)+t[z.index(max(z))]
print(th) ## Vth
print(max(z)**-1) ## s.s
```
