```
read = '/content/drive/My Drive/Labwork/as depo.csv'
x = pd.read_csv(read,engine='python',header=2)

gateV = pd.DataFrame(x, columns=['Gate Vg (V)'])
drainI = pd.DataFrame(x, columns=['Drain Id (A)'])
k=np.array(drainI)
o_depo=max(k)/min(k) ## on/off current ration
y=np.array(drainI)
z=[]
for i in range(len(y)-1):
  z.append(2*(y[i]-y[i-1]))
t=np.array(gateV)
th_depo=-y[z.index(max(z))]/max(z)+t[z.index(max(z))] ## Vth
l=np.log10(y)
q=[]
for i in range(len(y)-1):
  q.append(2*(l[i]-l[i-1]))
c_depo=(max(q))**-1 ## s.s
```

```
x=['as depo','200','250','300'] ## S.S
y=[c_depo,c_200,c_250,c_300]
plt.plot(x,y)
plt.scatter(x,y)
plt.xlabel("Annealing temperture (°C)")
plt.ylabel("S.S (V/decade)")
plt.show()

x=['as depo','200','250','300'] ## On/Off
y=[o_depo,o_200,o_250,o_300]
plt.plot(x,y)
plt.scatter(x,y)
plt.xlabel("Annealing temperture (°C)")
plt.ylabel("On/Off current ratio")
plt.show()

x=['as depo','200','250','300'] ## On/Off
y=[th_depo,th_200,th_250,th_300]
plt.plot(x,y)
plt.scatter(x,y)
plt.xlabel("Annealing temperture (°C)")
plt.ylabel("Threshold Voltage (V)")
plt.show()
```
